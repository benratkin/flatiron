## The Scenario

You've been tasked with designing an experiment to test whether a new email template will be more effective for your company's marketing team. The current template has a 5% response rate (with standard deviation .0475), which has outperformed numerous other templates in the past. The company is excited to test the new design that was developed internally but nervous about losing sales if it is not to work out. As a result, they are looking to determine how many individuals they will need to serve the new email template in order to detect a 1% performance increase (or decrease).


## Step 1: State the Null Hypothesis, $H_0$

```python
from statsmodels.stats.power import TTestIndPower
import matplotlib.pyplot as plt
import numpy as np
```

State your null hypothesis here (be sure to make it quantitative as before)

```python
# H_0 = response rate doesn't change between templates
```

## Step 2: State the Alternative Hypothesis, $H_1$

State your alternative hypothesis here (be sure to make it quantitative as before)

H_1 = Response rate $temp_1$ > $temp_0$

## Step 3: Calculate n for standard alpha and power thresholds

Now define what $\alpha$ and $\beta$ you believe might be appropriate for this scenario.
To start, arbitrarily set $\alpha$ to 0.05. From this, calculate the required sample size to detect a .01 response rate difference at a power of .8.

> Note: Be sure to calculate a normalized effect size using Cohen's d from the raw response rate difference.

```python
# Calculate the required sample size

power = TTestIndPower()
power.solve_power(effect_size=.01/.0475, alpha=.05, power=.8, alternative='larger')
```

## Step 4: Plot Power Curves for Alternative Experiment Formulations

While you now know how many observations you need in order to run a t-test for the given formulation above, it is worth exploring what sample sizes would be required for alternative test formulations. For example, how much does the required sample size increase if you put the more stringent criteria of $\alpha=.01$? Or what is the sample size required to detect a .03 response rate difference at the same $\alpha$ and power thresholds? To investigate this, plot power vs sample size curves for alpha values of .01, .05 and .1 along with varying response rate differences of .005, .01, .02 and .03.

```python
#Your code; plot power curves for the various alpha and effect size combinations
xs_alpha = []
ys_alpha = []
for alpha in [.01, .05, .1]:
    power = TTestIndPower()
    xs_alpha.append(alpha)
    n = power.solve_power(effect_size=.01/.0475, alpha=alpha, power=.8, alternative='larger')
    ys_alpha.append(n)
plt.xlabel('Alpha')
plt.ylabel('N required')
plt.plot(xs_alpha, ys_alpha);
        
```

```python
xs_mean_diff = []
ys_mean_diff = []
for mean_diff in [.005, .01, .02, .03]:
    power = TTestIndPower()
    xs_mean_diff.append(mean_diff)
    n = power.solve_power(effect_size=mean_diff/.0475, alpha=.05, power=.8, alternative='larger')
    ys_mean_diff.append(n)
plt.xlabel('mean_diff')
plt.ylabel('N required')
plt.plot(xs_mean_diff, ys_mean_diff);
```

```python
xs_power = []
ys_power = []
for power in [.8, .85, .9, .95]:
    power_obj = TTestIndPower()
    xs_power.append(power)
    n = power_obj.solve_power(effect_size=.01/.0475, alpha=.05, power=power, alternative='larger')
    ys_power.append(n)
plt.xlabel('mean_diff')
plt.ylabel('N required')
plt.plot(xs_power, ys_power);
```

## Step 5: Propose a Final Experimental Design

Finally, now that you've explored some of the various sample sizes required for statistical tests of varying power, effect size and type I errors, propose an experimental design to pitch to your boss and some of the accompanying advantages or disadvantages with it.

### Your answer here

In my opinion, a type II error (incorrectly concluding that the new template is more effective) is more costly in this scenario and should be avoided. We would incur the cost of switching our existing template and systems with no increase in response rate. An alpha level of .05 will be sufficient to accomplish this goal but could be decreased.

That said, we do want to detect a positive change if there is one to detect (by increasing the power). Because email sample sizes are easily obtained I suggest we try and maximize our power by using a large number of samples. I suggest a power of .95 which will require a minimum sample size of ~450.

Additionally, even a small increase in response rate is significant. For example, assuming our existing response rate of .06 an increase to .065 would be an overall 8.3% increase in response rate just by changing the email template. This would require a number of samples closer to 1000.

In conclusion, I suggest we conduct an experiment that is able to detect effect sizes of .005 or greater with a power of .95 which will require a sample size of N ~ 1000. This test will have an alpha (rate of type I errors) of .05.
